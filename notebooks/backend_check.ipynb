```python
from qiskit_ibm_runtime import fake_provider
import inspect

def list_available_fake_backends():
    print(f"{'Backend Name':<22} | {'Qubits':<10} | {'Description'}")
    print("-" * 50)
    
    # fake_provider 모듈 내의 모든 클래스를 검사
    for name, cls in inspect.getmembers(fake_provider, inspect.isclass):
        # 'Fake'로 시작하고 'V2'로 끝나는 백엔드 (최신 인터페이스) 위주로 확인
        # 혹은 단순히 'Fake'로 시작하는 모든 백엔드 확인
        if name.startswith("Fake") and "V2" not in name: 
            # 인스턴스화하여 속성 확인 (시간이 조금 걸릴 수 있음)
            try:
                backend = cls()
                num_qubits = backend.num_qubits
                print(f"{name:<22} | {num_qubits:<10} | {backend.version}")
            except Exception:
                continue

if __name__ == "__main__":
    list_available_fake_backends()
```
